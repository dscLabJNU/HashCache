```python
import pandas as pd 

log_dir="../logs"
faascache="FaaSCache"
hashcache="HashCache"
openwhisk="OpenWhisk"
FaaSCache_data = pd.read_csv(f"{log_dir}/{faascache}/{faascache}_stats.csv")
HashCache_data = pd.read_csv(f"{log_dir}/{hashcache}/{hashcache}_stats.csv")
OpenWhisk_data = pd.read_csv(f"{log_dir}/{openwhisk}/{openwhisk}_stats.csv")

OpenWhisk_data
```

```python
API_PREFIX="/api/23bc46b1-71f6-4ed5-8c54-816aa4f8c502/"

evaluate_apis = [
    "/api/23bc46b1-71f6-4ed5-8c54-816aa4f8c502/tickets/getLeftTripTickets",
    "/api/23bc46b1-71f6-4ed5-8c54-816aa4f8c502/tickets/preserveTickets",
    "/api/23bc46b1-71f6-4ed5-8c54-816aa4f8c502/orders/queryOrdersForRefresh",
    "/api/23bc46b1-71f6-4ed5-8c54-816aa4f8c502/orders/payForTheOrder",
    "/api/23bc46b1-71f6-4ed5-8c54-816aa4f8c502/contacts/findContactsByAccountId/4d2a46c7-71cb-4cf1-b5bb-b68406d9da6f"
]
metrics = ["Average Response Time", "Request Count", "50%", '75%', '90%', '95%', '99%', '99.9%']
for api in evaluate_apis:
    short_api=api.replace(API_PREFIX, "")
    api_interface = short_api.split("/")[1]
    print(f"===== {api_interface} =====")
    api_hashcache = HashCache_data[HashCache_data['Name']==api]
    api_faascache = FaaSCache_data[FaaSCache_data['Name']==api]
    api_openwhisk = OpenWhisk_data[OpenWhisk_data['Name']==api]
    for metric in metrics:
        print(f"{metric}-latency:\n {api_hashcache[metric].values}, {api_faascache[metric].values}, {api_openwhisk[metric].values}")
```
